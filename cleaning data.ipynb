```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#df = pd.read_csv('file_location/file_name')
df = pd.read_csv('Complete_aljazeera.csv')
```

```python
#Remove Digits

df = df[~df.word.str.contains("\\d")]
```

```python
#Sort by decending frequency

df = df.sort_values('freq', ascending=False)
df.head(20)
```

```python
#Remove General Phrases
#You can choose whatever words you want to from the .head(20) above

df = df[~df.word.str.contains("people")]
df = df[~df.word.str.contains("day")]
df = df[~df.word.str.contains("continue")]
df = df[~df.word.str.contains("day")]
df = df[~df.word.str.contains("city")]
```

```python
#Remove Dups
df = df.drop_duplicates("word")
```

```python
#combine similar word frequency

corona = df[df.word.str.contains('corona')]
covid = df[df.word.str.contains('covid')]
corona_freq = corona.freq.sum() + covid.freq.sum()

vaccine = df[df.word.str.contains('vaccin')]
vaccine_freq = vaccine.freq.sum()
```

```python
#slice terms and adjust freq

temp_corona = df.loc[df['word'] == 'corona']
temp_corona.freq = corona_freq

df = df[~df.word.str.contains("corona")]
df = df[~df.word.str.contains("covid")]

temp_vaccine = df.loc[df['word'] == 'vaccin']
temp_vaccine.freq = vaccine_freq

df = df[~df.word.str.contains("vaccin")]
```

```python
#Append Slice

df = df.append(temp_corona)
df = df.append(temp_vaccine)
print(df[df.word.str.contains('corona')])
print(df[df.word.str.contains('vaccin')])
```

```python
#Sort again

df = df.sort_values('freq', ascending=False)
```
