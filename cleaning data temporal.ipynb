```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#df = pd.read_csv('file_location/file_name')
df = pd.read_csv('msnbc.csv')
#df = pd.read_csv('Complete_aljazeera.csv')
```

```python
#drop nan values

df = df[df["word"].notna()]
df = df[df["freq"].notna()]
df = df[df["date"].notna()]
print(sum(df.word.isna()) + sum(df.freq.isna()) + sum(df.date.isna()))
```

```python
#Remove Digits

df = df[~df.word.str.contains("\\d")]
```

```python
#Sort by decending frequency

df = df.sort_values('freq', ascending=False)
df.head(20)
```

```python
#Remove General Phrases
#You can choose whatever words you want to from the .head(20) above

df = df[~df.word.str.contains("mr")]
df = df[~df.word.str.contains("yes")]
df = df[~df.word.str.contains("gentleman")]
df = df[~df.word.str.contains("people")]
df = df[~df.word.str.contains("day")]
df = df[~df.word.str.contains("continue")]
df = df[~df.word.str.contains("day")]
df = df[~df.word.str.contains("city")]
```

```python
#Remove Dups
#df = df.drop_duplicates("word")
```

```python
#combine similar word frequency

corona = df[df.word.str.contains('corona')]
corona = corona.append(df[df.word.str.contains('covid')])
#corona_freq = corona.freq.sum() + covid.freq.sum()

vaccine = df[df.word.str.contains('vacc')]
vaccine = vaccine.append(df[df.word.str.contains('vax')])

anti = vaccine[vaccine.word.str.contains('anti')]
anti = anti.append(vaccine[vaccine.word.str.contains('no')])
vaccine = vaccine[~vaccine.word.str.contains("anti")]
vaccine = vaccine[~vaccine.word.str.contains("no")]
#vaccine_freq = vaccine.freq.sum()

mask = df[df.word.str.contains('mask')]
mask = mask.append(df[df.word.str.contains('N95')])
mask = mask.append(df[df.word.str.contains('covering')])

#corona
```
